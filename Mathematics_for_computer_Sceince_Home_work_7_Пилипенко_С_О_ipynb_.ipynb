---
Розв'язання:

$$
P(X = 2) = \frac{A \cdot B}{C}

A = C_5^2 = \frac{5!}{2! \cdot 3!}

$$
B = C_{45}^4 = \frac{45!}{4! \cdot 41!}

$$

## Завдання 2

В тирі стрілець проводить 7 пострілів по мішені з ймовірністю влучання кожного 0.8. Яка ймовірність того, що буде не менше за 5 влучань? Розв'яжи аналітично.

---
Розв'язання:

 **Формула ймовірності біноміального розподілу**

$$
P(X = k) = C_n^k \cdot p^k \cdot q^{n - k}
$$

---

 **Знайдемо ймовірність того, що $X \geq 5$**

$$
P(X \geq 5) = P(5) + P(6) + P(7)
$$

---
** Обчислюємо кожну ймовірність**

🔹 Для $P(5)$:

$$
C_7^5 = \frac{7!}{5! \cdot 2!} = \frac{5040}{120 \cdot 2} = 21
$$

$$
P(5) = 21 \cdot 0{,}8^5 \cdot 0{,}2^2 = 21 \cdot 0{,}32768 \cdot 0{,}04 = 0{,}275
$$

🔹 Для $P(6)$:

$$
C_7^6 = \frac{7!}{6! \cdot 1!} = \frac{5040}{720 \cdot 1} = 7
$$

$$
P(6) = 7 \cdot 0{,}8^6 \cdot 0{,}2 = 7 \cdot 0{,}262144 \cdot 0{,}2 = 0{,}367
$$

🔹 Для $P(7)$:

$$
C_7^7 = 1
$$

$$
P(7) = 1 \cdot 0{,}8^7 = 0{,}2097152 \approx 0{,}210
$$

---
**Підсумовуємо:**

$$
P(X \geq 5) = P(5) + P(6) + P(7) = 0{,}275 + 0{,}367 + 0{,}210 = \boxed{0{,}852}
$$

---

 **Висновок:**  
Імовірність того, що влучень буде не менше 5 — **приблизно 0,852**.

_"Оскільки події взаємовиключні, обчислюємо суму ймовірностей."_


## Завдання 3

Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12?  
Розв'яжи програмно, згенерувавши всі можливі комбінації.

---
Розв'язання:

````python
import itertools

sums = [sum(p) for p in itertools.product(range(1, 7), repeat=3)]
count_11 = sums.count(11)
count_12 = sums.count(12)

print(f"Ймовірність 11: {count_11}/216 = {count_11/216:.3f}")
print(f"Ймовірність 12: {count_12}/216 = {count_12/216:.3f}")
````

## Завдання 6
Урна містить одну кулю, про яку відомо, що вона або біла, або чорна з однаковими ймовірностями. В урну кладуть білу кулю і потім навмання виймають одну кулю. Вона виявилася білою. Яка ймовірність того, що куля, яка залишилася, є білою? Розв'яжи аналітично  

---
Розв'язання:

### Позначимо:

- $H_1$: перша куля — **біла**  
- $H_2$: перша куля — **чорна**  
- $B$: витягли білу

$$
P(H_1) = P(H_2) = 0.5
$$

### Умовні ймовірності:

$$
P(B|H_1) = \frac{2}{2} = 1
$$

$$
P(B|H_2) = \frac{1}{2}
$$

### З формули Байєса:

$$
P(H_1|B) = \frac{P(B|H_1) \cdot P(H_1)}{P(B|H_1)P(H_1) + P(B|H_2)P(H_2)} = \frac{1 \cdot 0.5}{1 \cdot 0.5 + 0.5 \cdot 0.5} = \frac{0.5}{0.75} = \frac{2}{3}
$$

Отже, **ймовірність того, що залишилася біла куля = 2/3.**



## Завдання 7. Випадкове блукання
Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1. В початковий момент часу $t=0$ ціна рівна 0. Визнач середню ціну акції через 3 одиниці часу ($t=3$).  
а) Виріши аналітично.

б) Напиши симуляцію данного процесу. Порахуй середнє значення та намалюй гістограму ціни для 10, 100, 1000 та 10000 симуляцій.

---
Розв'язання:

Маємо випадковий процес: кожен крок змінює ціну на +2 або -1 з однаковою ймовірністю (0.5).

Позначимо $X_i$ — зміна на $i$-му кроці:
- $X_i = +2$ з імовірністю 0.5  
- $X_i = -1$ з імовірністю 0.5

### Математичне сподівання одного кроку:
$$
E(X_i) = 2 \cdot 0.5 + (-1) \cdot 0.5 = 0.5
$$

### Через 3 кроки:
$$
E(S) = 3 \cdot E(X_i) = 3 \cdot 0.5 = 1.5
$$

Отже, аналітично очікувана ціна = **1.5**


````python
import numpy as np
def stock_price_at_time(t):
    steps = np.random.choice([2, -1], size=t, p=[0.5, 0.5])
    return np.sum(steps)

def simulate_n_times(n: int, t: int) -> list:
    np.random.seed(42)
    return [stock_price_at_time(t) for _ in range(n)]



````

````python
import matplotlib.pyplot as plt
for n in [10, 100, 1000, 10000]:
    results = simulate_n_times(n, 3)
    mean_val = np.mean(results)
    plt.hist(results, bins=range(min(results), max(results)+1), alpha=0.5, label=f'n={n}')
    print(f"Середнє значення для {n} симуляцій: {mean_val:.2f}")

plt.title("Гістограма цін після 3 кроків")
plt.xlabel("Ціна")
plt.ylabel("Частота")
plt.legend()
plt.grid(True)
plt.show()
````
